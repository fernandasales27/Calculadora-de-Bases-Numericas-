```python
print("-----calculadora de bases-----")
print("selecione a opção")
print("1-binario para decimal ")
print("2-decimal para binario")
print("3-binario para hexadecimal ")
print("4-hexadecimal para binario ")
print("5-binario para octal")
print("6-octal para binario")
print("7-decimal para octal")
print("8-octal para decimal")
print("9-decimal para hexadecimal")
print("10-hexadecimal para decimal")
print("11-octal para hexadecimal")
print("12-hexadecimal para octal")
print("13-sair")

code=int(input('digite a opção: '))


while True:
  #code=int(input('digite a opção: '))

  if code==1:
     #1-binario para decimal
      num=str(input('digite numero a ser convertido:'))
      binario=int(num,2)
      print('|',num, "em decimal é= ", binario,'|')
  elif code==2:
    #2-decimal para binario
    num=int(input('digite numero a ser convertido:'))
    print('|',num, "em binario é=", bin(num),'|')

  elif code==3:
    #3-binario para hexadecimal
    num=str(input('digite numero a ser convertido:'))
    num_decimal=int(num,2)
    num_hex=hex(num_decimal)
    print('|',num, 'em hexadecimal é=', num_hex,'|')

  elif code==4:
    #4-hexadecimal para binario
    num=str(input('digite numero a ser convertido:'))
    num_decimal = int(num, 16)
    num_binario = bin(num_decimal)
    print('|',num, 'em binario é=', num_binario,'|')

  elif code==5:
    #5-binario para octal
   num=str(input('digite numero a ser convertido:'))
   num_decimal = int(num, 2)
   num_octal = oct(num_decimal)
   print('|',num,'em octal é=',num_octal,'|')

  elif code==6:
  #6-octal para binario
   num=str(input('digite numero a ser convertido:'))
   num_decimal = int(num, 8)
   num_binario = bin(num_decimal)
   print('|',num, 'em num_binario',num_binario,'|')

  elif code==7:
    #7-decimal para octal
   num=int(input('digite numero a ser convertido:'))
   num_octal=oct(num)
   print('|',num,'em octal é=',num_octal,'|')

  elif code==8:
    #8-octal para decimal
   num=str(input('digite numero a ser convertido:'))
   num_decimal = int(num, 8)
   print('|',num,'em decimal é=',num_decimal,'|')

  elif code==9:
    #9-decimal para hexadecimal
   num=int(input('digite numero a ser convertido:'))
   num_hex = hex(num)
   print('|',num,'em hexadecimal é =',num_hex,'|')

  elif code==10:
    #10-hexadecimal para decimal
   num=int(input('digite numero a ser convertido:'))
   num_decimal = int(num, 16)
   print('|',num,'em decimal é=',num_decimal,'|')

  elif code==11:
    #11-octal para hexadecimal
   num=str(input('digite numero a ser convertido:'))
   num_decimal = int(num, 16)
   num_octal = oct(num_decimal)
   print('|',num, 'em hexadecimal é=', num_octal,'|')

  elif code==12:
    #12-hexadecimal para octal
   num=str(input('digite numero a ser convertido:'))
   num_decimal = int(num, 16)
   num_octal = oct(num_decimal)
   print('|',num,'em ocatal é=',num_octal,'|')

  elif code== 13:
    break

  else:
    print("opção inválida, tente novamente")
  code=int(input('digite a opção: '))
```
